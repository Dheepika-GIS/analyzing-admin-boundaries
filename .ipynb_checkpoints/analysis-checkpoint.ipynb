# Challenge Exercise: Road Density Analysis by District

Objective: Calculate the road density of National Highways per district (i.e., total length of NH roads per square kilometer of each district).

### Reading Spatial Data

Read the layers: Load both the roads and districts layers from your GPKG.

```python
import os

data_pkg_path = 'data'
filename = 'karnataka.gpkg'
path = os.path.join(data_pkg_path, filename)
```

```python
roads = gpd.read_file(path, layer='karnataka_major_roads')
districts = gpd.read_file(path, layer='karnataka_districts')
```

### Filtering Data

Filter: Keep only National Highways from the roads layer.

```python
NH_Filtered = roads[roads['ref'].str.match('^NH') == True]
#NH_Filtered.head()
```

### Reprojecting Layers

Project both layers to a projected CRS suitable for the area (so that area and length are meaningful, e.g., UTM).

```python
roads_reprojected = NH_Filtered.to_crs('EPSG:32643')
districts_reprojected = districts.to_crs('EPSG:32643')

print(roads_reprojected.crs)
print(districts_reprojected.crs)
```

### Calculating Area

Calculate area for each district (in square kilometers).

```python
districts_reprojected['area_km2'] = districts_reprojected['geometry'].area / 1_000_000
districts_reprojected['area_km2'] = districts_reprojected['area_km2'].round(2)
```

## Spatial join

Spatial join: Join roads to districts using intersects.

```python
joined = gpd.sjoin(districts_reprojected, roads_reprojected, how='left', predicate='intersects')
```

### Calculating Length

Calculate the length of each joined road segment

```python
joined["length_km"] = joined.geometry.length / 1000
```

Group by district identifier (like district name or ID), and sum the road lengths:

```python
NH_length_per_district = joined.groupby('DISTRICT')['length_km'].sum().reset_index()
```

Merge with the original districts_reprojected, which already has area_km2:

```python
result = districts_reprojected.merge(NH_length_per_district, on='DISTRICT', how='left')
```

### Calculate Road Density

```python
result['road_density_km_per_km2'] = result['length_km'] / result['area_km2']
```

```python
Road_density_csv = result

csv_file = 'Road_density_per_square_km.csv'
csv_path = os.path.join(data_pkg_path, csv_file)

Road_density_csv.to_csv(csv_path, index=False)

print(f"Saved CSV file as: {csv_path}")
```
